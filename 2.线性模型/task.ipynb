```python
import numpy as np
import matplotlib.pyplot as plt

# 设y = 2x + 1
x_data = [1.0, 2.0, 3.0]
y_data = [3.0, 5.0, 7.0]

def forward(w, b, x):
    return w * x + b

def loss(y_hat, y):
    return (y_hat - y) ** 2

w_coverage = np.arange(0.0, 4.0, 0.1)
b_coverage = np.arange(0.0, 4.0, 0.1)

w, b = np.meshgrid(w_coverage, b_coverage)
mse = np.zeros(w.shape)

for x_val, y_val in zip(x_data, y_data):
    y_hat_val = forward(w, b, x_val)
    loss_val = loss(y_hat_val, y_val)
    mse += loss_val

fig = plt.figure()

ax = plt.axes(projection='3d')
plt.xlabel(r'w', fontsize=20, color='cyan')
plt.ylabel(r'b', fontsize=20, color='cyan')

ax.plot_surface(w, b, mse, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
plt.show()
```
